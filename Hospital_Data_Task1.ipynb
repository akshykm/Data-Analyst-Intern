```python
import pandas as pd
from google.colab import drive
drive.mount('/content/gdrive')
df = pd.read_csv('gdrive/My Drive/KaggleV2-May-2016.csv')
display(df.head(10))
display(df.sample(3))
```

```python
# Reset the index and start from 1
df.index = range(1, len(df) + 1)
```

```python
# Step 1:  Identify and handle missing values
missing = df.isnull().sum()
print(missing)
```

```python
# Step 2: Find number of duplicate rows
duplicate_count = df.duplicated().sum()
print(f"Total duplicate rows: {duplicate_count}")
```

```python
# Step 3:  Standardize text values
df['gender'] = df['gender'].str.strip().str.lower().map({
    'male': 'Male',
    'm': 'Male',
    'female': 'Female',
    'f': 'Female'
})
```

```python
# Step 4:   Convert date formats
df['scheduledday'] = pd.to_datetime(df['scheduledday']).dt.strftime('%d-%m-%Y')
df['appointmentday'] = pd.to_datetime(df['appointmentday']).dt.strftime('%d-%m-%Y')
```

```python
# Step 5: Rename column headers
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# Step 6:  Check and fix data types
df['age'] = df['age'].astype(int)
df['patientid'] = df['patientid'].astype('int64')
df['appointmentid'] = df['appointmentid'].astype('int')
```
